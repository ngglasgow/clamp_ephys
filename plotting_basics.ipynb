```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
# prepare some data
x = np.linspace(0, 10, 100)
y = np.cos(x)
y2 = x**2
y3 = x*2

# make a figure with multiple subplots
# sharex=True links the pan and zoom on x
fig, axs = plt.subplots(nrows=3, figsize=(4, 6), sharex=True, constrained_layout=True)
axs[0].plot(x, y)
axs[1].plot(x, y2)
axs[2].plot(x, y3)
```

```python
# for bokeh you'll need to install via: 
# !conda install bokeh
# for quick options check: https://docs.bokeh.org/en/latest/docs/user_guide/quickstart.html
from bokeh.layouts import gridplot
from bokeh.plotting import figure
from bokeh.io import output_notebook, show
output_notebook()
```

```python
# prepare some data
x = np.linspace(0, 10, 100)
y = np.cos(x)
y2 = x**2
y3 = x*2

# plot params
width = 600
height = 250

# create a new plot
s1 = figure(width=width, height=height)
s1.line(x, y, line_width=2)

# NEW: create a new plot and share x range
s2 = figure(width=width, height=height, x_range=s1.x_range)
s2.line(x, y2, line_width=2, color="firebrick")

# NEW: create a new plot and share only one range
s3 = figure(width=width, height=height, x_range=s1.x_range, x_axis_label='Time (ms)')
s3.line(x, y3, line_width=2, color="olive")

# NEW: put the subplots in a gridplot
p = gridplot([[s1], [s2], [s3]])

# show the results
show(p)
```
